```python
lista_de_numeros = [12,124,36,14,21,36]

for n in lista_de_numeros:
  if n == 36:
    print(f"Has encontrado el numero")
  else:
    print(f"seguimos buscando, n = {n}")
```

# Ahora usaremos break
## Sirve para detener el bucle

```python
lista_de_numeros = [12,124,36,14,21,36]

for n in lista_de_numeros:
  if n == 36:
    print(f"Has encontrado el numero")
    break
  else:
    print(f"seguimos buscando, n = {n}")
```
